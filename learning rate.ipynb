```python
import random
import matplotlib.pyplot as plt
import numpy as np
 
# setting the x - coordinates
x = np.arange(0, 10, 1)
# setting the corresponding y - coordinates
y = x
 
# potting the points
plt.plot(x, y)
 
# function to show the plot
plt.show()


error=[]
q=[]

def calErr(theta):
    h=theta*x
    j=0

    for i in range (0,4):
       j+=(h[i]-y[i])**2
    j/=3
    return j



learningRate=20

start=100
i=start

j=0
check=0


while(True):
    q.append(i)
    pt=calErr(q[j])
    error.append(pt)
    j=j+1
    i=i+learningRate
    
    if(j==3):
        if(error[0]>error[1]):
            check=1
    
    if(j>2):
        if((error[j-2])<(error[j-1])):
            if(check==1):
                break
            elif(check==0):
                learningRate=-learningRate
                check=1
                i=start
                
    if(j>4):
        if(((error[j-4])-(error[j-3]))>0.80*((error[j-2])-(error[j-1])) and ((error[j-2])-(error[j-1]))>1):
            learningRate=learningRate/1.5
            
            
                

plt.plot(q, error)
plt.show()
```
